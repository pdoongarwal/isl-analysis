```python
#import packages
import pandas as pd
import seaborn as sns
```

```python
YEARS = list(range(2014, 2021))
```

### Clean and wrangle data
- Currently our data is divided for each season. We need to combine it into single dataframe to perform overall analysis. So, we will add a new column to represent season.
- We will add another column to show if a team made it into palyoffs or not.

```python
def read_data():
    """
    Read data for each year.
    
    Returns:
    dict: A dictonary with year to that year's data mapping
    """
    df_years = {}
    for year in YEARS:
        df = pd.read_csv('./data/{}/team/combined.csv'.format(year))
        df_years[year] = df
        
    return df_years
```

```python
def add_columns(df_years):
    """
    Add columns for each year's dataframe.
    
    Parameters:
    df_years (dict): A dictonary with year to that year's data mapping
    
    Returns:
    dict: A dictonary with year to that year's data mapping
    """
    updated_df = df_years.copy()
    for year in YEARS:
        df = updated_df[year]
        df = df.sort_values(by='match_played', ignore_index=True, ascending=False)
        df['made_playoff'] = pd.Series([1]*4 + [0]*(len(df) - 4))
        df['year'] = year
        updated_df[year] = df
        
    return updated_df
```

```python
def combine_data(df_years):
    """
    Combine each year's data in a single dataframe.
    
    Parameters:
    df_years (dict): A dictonary with year to that year's data mapping
    
    Return:
    Dataframe: A combined dataframe for all seasons data
    """
    df_values = []
    for year in YEARS:
        df_values.append(df_years[year])
        
    return pd.concat(df_values, ignore_index=True)
```

```python
df_years = read_data()
df_years = add_columns(df_years)
df = combine_data(df_years)
df.head()
```

### Normalize Data
Over season teams played different number of matches as number of teams in each season were different.
It would be better if consider stats in terms of per match. We will simply divide game stats by matches played

**NOTE: Even in same season teams played different number of matches as this data included stats for playoffs as well.**

```python
def normalize_data(df):
    """
    Normalize game stats by games played.
    
    Parameters:
    df (Dataframe): Game stats for each team for each year
    
    Return:
    Dataframe: Normalized game stats for each team for each year
    """
    columns_to_normalize = ['win', 'draw', 'loss', 'goal', 'shot', 'cross', 'pass', 'touch', 'cleansheet', 'save', 'tackle', 'foul', 'redcard', 'yellowcard']
    normalized_df = df.copy()
    for column in columns_to_normalize:
        normalized_df[column] = normalized_df[column] / normalized_df['match_played']
        
    return normalized_df
```

```python
clean_df = df.fillna(0)
normalized_df = normalize_data(clean_df)
```

### Analyze factors separating playoff and non-playoff teams

#### Points to note:

- A significant difference in goals per game.
- Playoff teams have more shots per game but fewer crosses.
