```python
import json
import pandas as pd
```

```python
YEAR = 2020
files = ['win', 'draw', 'loss', 'goal', 'shot',  'cross', 'pass', 'touch', 'cleansheet', 'save', 'tackle', 'foul', 'redcard', 'yellowcard']
data_dict = {}

for file in files:
    with open('./data/{}/team/{}.json'.format(YEAR, file), 'r', encoding='utf8') as f:
        print(file)
        data_dict[file] = json.load(f)['data']
```

```python
for team in data_dict['goal']:
    print(team['team_name'], team['match_played'], team['value'], team['value']/team['match_played'])
```

```python
clean_data_dict = {}

for file in files:
    data = data_dict[file]
    
    for team_data in data:
        team_short_name = team_data['team_short_name']
        
        if team_short_name not in clean_data_dict:
            clean_data_dict[team_short_name] = {
                'team_name': team_data['team_name'],
                'team_short_name': team_short_name,
                'match_played': team_data['match_played']
            }
            
        clean_data_dict[team_short_name][file] = team_data['value']
```

```python
clean_data_dict
```

```python
clean_data = []
for value in clean_data_dict.values():
    clean_data.append(value)
```

```python
clean_df = pd.DataFrame(clean_data)
clean_df
```

```python
clean_df.fillna(0)
```

```python
clean_df.to_csv('./data/{}/team/combined.csv'.format(YEAR), index=False)
```

```python
clean_df['points'] = clean_df['win']*3 + clean_df['draw']
clean_df = clean_df.sort_values(by='points')
clean_df['in_playoff'] = pd.Series([1]*4 + [0]*(len(clean_df) - 4))
clean_df
```
